# Código para capturar tabelas de classificação ano a ano e transformar em CSV

```python
# Trazer bibliotecas
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Arquivo salvo como html
with open(r"HTML_Classificacao\25-26\NBB_25_avg.html", "r", encoding="latin-1") as f:
    conteudo = f.read()
```

```python
# Criação de parser
soup = BeautifulSoup(conteudo, "html.parser")
```

```python
# Pega todas as linhas da tabela
linhas = soup.find_all('tr')
```

```python
# Extrai o texto de cada célula
data = []
for linha in linhas:
    cols = linha.find_all(['td', 'th'])
    cols = [col.get_text(strip=True) for col in cols]
    data.append(cols)
```

```python
# Exibe a tabela selecionada
df = pd.DataFrame(data[1:], columns=data[0])
```

```python
df.head()
```

```python
# Conferencia de valores inicial
df.info()
```

```python
# Apagar coluna index da tabela origem
df.drop(columns= '#', inplace=True)
df.drop(axis = 1, index= 0, inplace=True)
```

```python
# Conferência de valores
df.info()
```

```python
# Conversão de tipos de dados
df.drop(columns=['GP', 'MPG', 'PF'], inplace=True)
for col in df.columns:
    if col not in ['Team', 'GP']:
        df[col]=df[col].astype(float)
df.info()
```

```python
df.head()
```

```python
# Arquivo salvo como html
with open(r"HTML_Classificacao\25-26\NBB_25_misc.html", "r", encoding="latin-1") as f:
    conteudo = f.read()
```

```python
# Pega todas as linhas da tabela
linhas = soup.find_all('tr')
```

```python
# Exibe a tabela selecionada
df3 = pd.DataFrame(data[1:], columns=data[0])
```

```python
df3.info()
```
